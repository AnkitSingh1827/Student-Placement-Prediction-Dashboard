```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.10,      # 10% for testing
    random_state=42,      # Reproducible results
    stratify=Y            # Keeps the class distribution the same in train and test
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# scikit-learn is already installed and imported above
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=42, max_iter=1000)
clf.fit(X_train, Y_train)
```

```python
from sklearn.decomposition import PCA

# Reduce to 2 dimensions
pca = PCA(n_components=2)

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf_pca = LogisticRegression(random_state=42)
clf_pca.fit(X_train_pca, Y_train)
```

```python
y_pred = clf_pca.predict(X_test_pca)
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(Y_test, y_pred)
```
